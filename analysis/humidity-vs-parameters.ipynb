```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def LinearVariation(x,y,data_label):
    plt.plot(x, y, color = 'r', linestyle = 'dashed',marker = 'o',label = data_label)
    plt.xticks(rotation = 25)
    plt.xlabel('Humidity')
    plt.ylabel(data_label)
    plt.title('Variation of Humidity with '+data_label, fontsize = 15)
    plt.grid()
    plt.legend()
    plt.show()
```

```python
# Taking Parameters
time_stamp = []
ph_value = []
luminous_intensity = []
humidity = []
temperature = []
soil_moisture = []
water_level = []
```

```python
### For initial growth
```

```python
data = pd.read_csv('../datasets/day1_8.csv')
day = data.iloc[:,0].values;
time_stamp = data.iloc[:,1].values;
indexed_time_stamp = day + time_stamp
ph_value = data.iloc[:,2].values
luminous_intensity = data.iloc[:,3].values
humidity = data.iloc[:,4].values
temperature = data.iloc[:,5].values
soil_moisture = data.iloc[:,6].values
water_level = data.iloc[:,7].values
```

```python
LinearVariation(humidity,ph_value,"PH value")
```

```python
LinearVariation(humidity,luminous_intensity,"Sunlight")
```

```python
LinearVariation(humidity,temperature,"Temperature")
```

```python
LinearVariation(humidity,soil_moisture,"Soil Moisture")
```

```python
LinearVariation(humidity,water_level,"Water Level")
```
