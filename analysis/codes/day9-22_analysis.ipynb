## Linear Variation with respect to parameters

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def timeSeriesAnalyse(x,y,data_label):
    plt.plot(x, y, color = 'g', linestyle = 'dashed',marker = 'o',label = data_label)
    plt.xticks(rotation = 25)
    plt.xlabel('Date-Time')
    plt.ylabel(data_label)
    plt.title('Time series analysis '+data_label, fontsize = 20)
    plt.grid()
    plt.legend()
    plt.show()

    
```

```python
# Taking Parameters
time_stamp = []
ph_value = []
luminous_intensity = []
humidity = []
temperature = []
soil_moisture = []
water_level = []
```

```python
# Reading data
data = pd.read_csv('../datasets/day9-22.csv')
day = data.iloc[:,0].values;
time_stamp = data.iloc[:,1].values;
indexed_time_stamp = day + time_stamp
ph_value = data.iloc[:,2].values
luminous_intensity = data.iloc[:,3].values
humidity = data.iloc[:,4].values
temperature = data.iloc[:,5].values
soil_moisture = data.iloc[:,6].values
water_level = data.iloc[:,7].values
```

```python
timeSeriesAnalyse(indexed_time_stamp,ph_value,"Ph Data")
```

```python
timeSeriesAnalyse(indexed_time_stamp,luminous_intensity,"Sun Light")
```

```python
timeSeriesAnalyse(indexed_time_stamp,humidity,"Humidity data")
```

```python
timeSeriesAnalyse(indexed_time_stamp,temperature,"Temperature data")
```

```python
timeSeriesAnalyse(indexed_time_stamp,water_level,"Water level data")
```

```python
timeSeriesAnalyse(indexed_time_stamp,soil_moisture,"Soil moisture data")
```

```python
# Finding range of values
print("Max and Min PH value:                    " , max(ph_value)," |",min(ph_value)) 
print("Max and Min Sunlight value:              " , max(luminous_intensity)," |",min(luminous_intensity)) 
print("Max and Min Humidity value:              " , max(humidity)," |",min(humidity)) 
print("Max and Min Temperature value:           " , max(temperature)," |",min(temperature)) 
print("Max and Min Soil Moisture value:         " , max(soil_moisture)," |",min(soil_moisture)) 
print("Max and Min Min value water level:       " , max(water_level)," |",min(water_level)) 
```
